### Play musical notes

```python
import numpy as np
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
    t = np.arange(1, sr * dur+1)/sr
    theta = t * 2*np.pi * freq + phase
    s = amp * np.sin(theta)
    return s, t, theta
```

```python
def getFrequency(note):
    A4=440
    notes = ['A', 'A#', 'B', 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#']

    octave = int(note[2]) if len(note) == 3 else int(note[1])
        
    keyNumber = notes.index(note[0:-1]);
    
    if (keyNumber < 3) :
        keyNumber = keyNumber + 12 + ((octave - 1) * 12) + 1; 
    else:
        keyNumber = keyNumber + ((octave - 1) * 12) + 1; 

    return A4 * 2** ((keyNumber- 49) / 12)
```

```python
amp = 1; sr = 10000; phase = 0

D4  = getFrequency('D4')
G4  = getFrequency('G4')
A4  = getFrequency('A4')
B4  = getFrequency('B4')
C5  = getFrequency('C5')
D5  = getFrequency('D5')
Eb5 = getFrequency('D#5')
E5  = getFrequency('E5')	 
F5  = getFrequency('F5')
G5  = getFrequency('G5')
A5  = getFrequency('A5')
Bb5 = getFrequency('A#5')
B5  = getFrequency('B5')
C6  = getFrequency('C6')
D6  = getFrequency('D6')
E6  = getFrequency('E6')
F6  = getFrequency('F6')
Gb6 = getFrequency('F#6')
G6  = getFrequency('G6')
```

```python
# The bridge on the river Kwai

notes = [D6, 0, D6, 0, D6, C6, B5, C6, D6, Gb6, G6, E6, Gb6, G6, D6, Gb6, G6,
         C6, 0, C6, B5, A5, B5, C6, D6, D5, G5, A5, B5, C6, B5, A5, G5, A5, B5, C6,
         D6, Gb6, 0, Gb6, E6, D6, B5, 0, B5, A5, G5, A5]

durs = [0.73, 0.02, 0.23, 0.02, 0.25, 0.25, 0.25, 0.25, 1.25, 0.25, 0.25, 0.5, 0.25,
        0.25, 0.5, 0.25, 0.25, 0.48, 0.02, 0.25, 0.25, 0.5, 0.25, 0.25, 0.5, 0.5, 0.5,
        0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.5, 0.25, 0.25, 0.5, 0.48, 0.02, 0.25, 0.25, 0.25, 
        0.23, 0.02, 0.5, 0.25, 0.25, 1]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```

```python
ipd.Audio(s_all, rate=sr)
```
